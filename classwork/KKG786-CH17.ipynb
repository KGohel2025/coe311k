```python
import numpy as np
import matplotlib.pyplot as plt

k = 1
hs = [.01, .02, .03, .05, .1, .2, .3, .5]
x_start = 0
x_end = 10
y_0 = 1.0
v_0 = 0.0

def exact_y(x):
    return np.cos(np.sqrt(k) * x)

def exact_v(x):
    return -np.sqrt(k) * np.sin(np.sqrt(k) * x)

def run_iter(h):
    x_values = np.arange(x_start, x_end + h, h)
    y_values = np.zeros(len(x_values))
    v_values = np.zeros(len(x_values))

    y_values[0] = y_0
    v_values[0] = v_0

    for i in range(1, len(x_values)):
        y_n = y_values[i - 1]
        v_n = v_values[i - 1]

        y_values[i] = y_n + h * v_n
        v_values[i] = v_n - h * k * y_n

    return x_values, y_values, v_values

plt.figure(figsize=(12, 6))

x_exact = np.linspace(x_start, x_end, 1000)
plt.plot(x_exact, exact_y(x_exact), linewidth=3, label='Exact y(x)')
plt.plot(x_exact, exact_v(x_exact), linewidth=3, linestyle='--', label='Exact v(x)')

for h in hs:
    x_values, y_values, v_values = run_iter(h)
    plt.plot(x_values, y_values, label=f'y (h={h})')
    plt.plot(x_values, v_values, linestyle='--', label=f'v (h={h})')

plt.xlabel('x')
plt.ylabel('Values')
plt.title("Euler's Method vs Exact Solution (y and v)")
plt.legend(loc="center left", bbox_to_anchor=(1, 0.5))
plt.grid(True)
plt.tight_layout()
plt.show()





"""

It does become unstable. 
Increasing k makes the system oscilate faster . The instability also shows up faster. 
When setting k = 10 h = 0.01 the solution works initially but starts deviating faster than when k=1.
Eulers method cant solve this with a larger h because of inaccuracy. 

"""
```
