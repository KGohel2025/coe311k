```python
import time
import numpy as np
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
            # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations



def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
# Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations




# Part A
 
A = np.array([[10, -1,  2,  0,  0],[-1, 11, -1,  3,  0],[ 2, -1, 10, -1,  0],[ 0,  3, -1,  8, -2],[ 0,  0,  0, -2,  9]])
b = np.array([14, 30, 26, 25, 37])

x0 = np.zeros_like(b)
tol = 1e-6
maxiterations = 100


#Jacobi
t0 = time.perf_counter()
solutionj, iterationsj = jacobi(A, b, x0, tol, max_iterations)
t1 = time.perf_counter()

residualj = b - A @ solutionj
resnormj = np.linalg.norm(residualj, ord=np.inf)

print("Jacobi")
print("Solution ", solutionj)
print("Number of Iterations ", iterationsj)
print("Time in seconds ", t1 - t0)
print("Residual ", resnormj)


# Gauss-Seidel 
t0 = time.perf_counter()
solutiongs, iterationsgs = gauss_seidel(A, b, x0, tol, maxiterations)
t1 = time.perf_counter()

residualgs = b - A @ solution_gs
resnormgs = np.linalg.norm(residualgs, ord=np.inf)

print("Gauss-Seidel")
print("Solution ", solutiongs)
print("Number of Iterations ", iterationsgs)
print("Time ", t1 - t0)
print("Residual ", resnormgs)


# the comparition
print("\nConvergence Comparison")
prin= ""
if iterationsgs < iterationsj :
    prin = "Gauss-Seidel"
else :
    prin = "Jacobi"
print("The one with fewer iterations is ", prin)
```
