```python
import time
import numpy as np
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
            # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations



def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
# Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations




# Part A
 
A = np.array([[10, -1,  2,  0,  0],[-1, 11, -1,  3,  0],[ 2, -1, 10, -1,  0],[ 0,  3, -1,  8, -2],[ 0,  0,  0, -2,  9]])
b = np.array([14, 30, 26, 25, 37])

x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
solution, iterations = jacobi(A, b, x0, tol, max_iterations)



x0 = np.zeros_like(b)
tol = 1e-6
max_iterations = 100
solution, iterations = gauss_seidel(A, b, x0, tol, max_iterations)


# ---- Jacobi timing + residual ----
t0 = time.perf_counter()
solution_j, iterations_j = jacobi(A, b, x0, tol, max_iterations)
t1 = time.perf_counter()

residual_j = b - A @ solution_j
resnorm_j = np.linalg.norm(residual_j, ord=np.inf)

print("Jacobi")
print("Solution ", solution_j)
print("Number of Iterations ", iterations_j)
print("Time in seconds ", t1 - t0)
print("Residual ", resnorm_j)


# ---- Gauss-Seidel timing + residual ----
t0 = time.perf_counter()
solution_gs, iterations_gs = gauss_seidel(A, b, x0, tol, max_iterations)
t1 = time.perf_counter()

residual_gs = b - A @ solution_gs
resnorm_gs = np.linalg.norm(residual_gs, ord=np.inf)

print("Gauss-Seidel")
print("Solution ", solution_gs)
print("Number of Iterations ", iterations_gs)
print("Time ", t1 - t0)
print("Residual ", resnorm_gs)


# the comparition
print("\nConvergence Comparison")
print("The one with few iterations is ", "Gauss-Seidel" if iterations_gs < iterations_j else "Jacobi")
```
