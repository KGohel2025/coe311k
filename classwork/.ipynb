```python
import numpy as np
import math as math
import matplotlib.pyplot as plt
import random as random
import matplotlib.patches as mpatches
def approximatePi(n):
    counter = 0
    for i in range(n):
        x = random.uniform(-1,1) 
        y = random.uniform(-1,1) 
        if (x * x + y * y < 1):
               counter = counter + 1
            
    return 4*counter/n


for i in range (0,100):
    approximate_pi = approximatePi(1000)
    truepi = np.pi
    true_error = abs(truepi - approximate_pi)
    relativeerror = abs((truepi-approximate_pi) /(truepi) )
    
    plt.plot(i, true_error, 'b.')
    plt.plot(i, relativeerror, 'r.' )
    
red_patch = mpatches.Patch(color='red', label='Relative Error')
blue_patch = mpatches.Patch(color='blue', label='Absolute Error')
plt.legend(handles=[blue_patch, red_patch])


plt.xlabel("Run number")
plt.ylabel("Error")
plt.title("Absolute vs Relative Error")
plt.show()
```
