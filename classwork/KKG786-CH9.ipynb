```python
import numpy as np
import matplotlib.pyplot as plt

#part 1


X = np.array([1, 2, 3, 4, 5])
Y = np.array([1.2, 2.8, 3.6, 4.5, 5.1])

A = np.vstack([X, np.ones(len(X))]).T
m, c = np.linalg.lstsq(A, Y, rcond=None)[0]

print(f"Slope: {m}")
print(f"Intercept: {c}")

plt.scatter(X, Y, label="Data Points")
plt.plot(X, m*X + c, 'g', label="Least Squares Fit")
plt.title("Original Data")
plt.legend()
plt.show()


#part2
rng = np.random.default_rng(seed=42)
Y_shuffled = rng.permutation(Y)

A2 = np.vstack([X, np.ones(len(X))]).T
m2, c2 = np.linalg.lstsq(A2, Y_shuffled, rcond=None)[0]

print(f"Slope after shuffle: {m2}")
print(f"Intercept after shuffle: {c2}")

plt.scatter(X, Y_shuffled, label="Shuffled Data")
plt.plot(X, m2*X + c2, 'r', label="Fit After Shuffle")
plt.title(" Shuffled Data")
plt.legend()
plt.show()


#part3

m_true = 2.5
b_true = -0.4

X3 = np.arange(0, 1.01, 0.01)
noise = rng.normal(0, 0.1, size=len(X3))
Y3 = m_true*X3 + b_true + noise

A3 = np.vstack([X3, np.ones(len(X3))]).T
m3, b3 = np.linalg.lstsq(A3, Y3, rcond=None)[0]

print(f"True slope: {m_true}, True intercept: {b_true}")
print(f"Fitted slope: {m3}")
print(f"Fitted intercept: {b3}")

plt.scatter(X3, Y3, s=10, label="Data")
plt.plot(X3, m_true*X3 + b_true, label="True Line")
plt.plot(X3, m3*X3 + b3, label="Least Squares Fit")
plt.title(" True Line vs Fitted Line")
plt.legend()
plt.show()
```
