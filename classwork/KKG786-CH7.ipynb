```python
# Part A
import time
import numpy as np
def jacobi(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = np.zeros_like(x)
        for i in range(n):
            s = sum(A[i][j] * x[j] for j in range(n) if j != i)
            x_new[i] = (b[i] - s) / A[i][i]
        if not np.isfinite(x_new).all():
            print("Jacobi diverged at iteration", k)
            return x_new, k
            # Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations



def gauss_seidel(A, b, x0, tol, max_iterations):
    n = len(b)
    x = x0.copy()
    for k in range(max_iterations):
        x_new = x.copy()
        for i in range(n):
            s1 = sum(A[i][j] * x_new[j] for j in range(i)) # Using already updated values
            s2 = sum(A[i][j] * x[j] for j in range(i + 1, n)) # Using old values
            x_new[i] = (b[i] - s1 - s2) / A[i][i]
        if not np.isfinite(x_new).all():
            print("Gauss seidel diverged at iteration", k)
            return x_new, k
# Check for convergence
        if np.linalg.norm(x_new - x, ord=np.inf) < tol:
            return x_new, k
        x = x_new
    return x, max_iterations




# Part B
A = np.array([
	[0.582745, 0.48	, 0.10	, 0.  	, 0.  	], 
 [0.48	, 1.044129, 0.46	, 0.10	, 0.  	],
 [0.10	, 0.46	, 1.10431 , 0.44	, 0.10	],
 [0.  	, 0.10	, 0.44	, 0.963889, 0.42	],
 [0.  	, 0.  	, 0.10	, 0.42	, 0.522565]], dtype=float)
b = np.array([1.162745, 2.084129, 2.20431 , 1.923889, 1.042565], dtype=float)

x0 = np.zeros_like(b, dtype = float)
tol = 1e-6
max_iterations = 100


#Jacobi
t0 = time.perf_counter()
solutionj, iterationsj = jacobi(A, b, x0, tol, max_iterations)
t1 = time.perf_counter()

residualj = b - A @ solutionj
resnormj = np.linalg.norm(residualj, ord=np.inf)

print("Jacobi")
print("Solution ", solutionj)
print("Number of Iterations ", iterationsj)
print("Time in seconds ", t1 - t0)
print("Residual ", resnormj)


# Gauss-Seidel 
t0 = time.perf_counter()
solutiongs, iterationsgs = gauss_seidel(A, b, x0, tol, max_iterations)
t1 = time.perf_counter()

residualgs = b - A @ solutiongs
resnormgs = np.linalg.norm(residualgs, ord=np.inf)

print("Gauss-Seidel")
print("Solution ", solutiongs)
print("Number of Iterations ", iterationsgs)
print("Time ", t1 - t0)
print("Residual ", resnormgs)

convergedj = iterationsj < max_iterations
convergedgs = iterationsgs < max_iterations


if not convergedj and not convergedgs:
    print("Neither method converged for Part B (matrix not diagonally dominant).")
elif convergedj and convergedgs:
    # the comparition

    print("\nConvergence Comparison")
    prin= ""
    if iterationsgs < iterationsj :
        prin = "Gauss-Seidel"
    else :
        prin = "Jacobi"
    print("The one with fewer iterations is ", prin)

else:
    print("Only one converged:", "Jacobi" if convergedj else "Gauss-Seidel")
    # the comparition

    print("\nConvergence Comparison")
    prin= ""
    if iterationsgs < iterationsj :
        prin = "Gauss-Seidel"
    else :
        prin = "Jacobi"
    print("The one with fewer iterations is ", prin)
```
