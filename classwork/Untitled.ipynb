```python
import numpy as np
```

```python
v1 = np.array([1,2,3])
v2 = np.array([4,5,6])
```

```python
print(v1)
print(v2)
```

```python
v_add = v1+v2
v_sup = v1-v2
v_sp = v1*v2
print(v_add)
print(v_sup)
print(v_sp)
```

```python
vscaled = v1*2
print(vscaled)
crossproduct = np.dot(v1,v2)
print(crossproduct)
magnitude = np.linalg.norm(v1)
print(magnitude)
```
