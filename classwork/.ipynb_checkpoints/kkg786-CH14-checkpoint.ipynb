```python
import numpy as np
import matplotlib.pyplot as plt

f = lambda x: np.sin(x)
a, b = 0, 2 * np.pi
N = 50
x = np.linspace(a, b, N)
# Step size

h = (b - a) / (N - 1)

 

# Central Difference Approximation for the derivative

df_dx = (f(x + h) - f(x - h)) / (2 * h)

# Forward Difference Approximation
### YOUR CODE HERE ###

forward = (f(x + h) - f(x)) / h

# Backward Difference Approximation

### YOUR CODE HERE

 
backward = (f(x) - f(x - h)) / h
# Plot the approximation and the true derivative

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, np.cos(x), label='Exact')
plt.plot(x, np.cos(x), '--', label='Exact')  # same (just for consistency)
plt.title('cos(x)')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.legend()

plt.subplot(2, 2, 2)
plt.plot(x, np.cos(x), label='Exact')
plt.plot(x, forward, '--', label='Forward')
plt.title('Forward Difference')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.legend()

plt.subplot(2, 2, 3)
plt.plot(x, np.cos(x), label='Exact')
plt.plot(x, backward, '--', label='Backward')
plt.title('Backward Difference')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.legend()

plt.subplot(2, 2, 4)
plt.plot(x, np.cos(x), label='Exact')
plt.plot(x, df_dx, '--', label='Central')
plt.title('Central Difference')
plt.xlabel('x')
plt.ylabel("f'(x)")
plt.legend()

plt.tight_layout()
plt.show()

## Do a canvas of 4 subplots or 4 graphs on one plot comparing cos(x), Forward Difference, Backwards Difference, and CentralDifference Methods

plt.figure(figsize=(10, 8))

plt.subplot(2, 2, 1)
plt.plot(x, np.cos(x))
plt.title('cos(x)')
plt.xlabel('x')
plt.ylabel("f'(x)")

plt.subplot(2, 2, 2)
plt.plot(x, forward)
plt.title('Forward Difference')
plt.xlabel('x')
plt.ylabel("f'(x)")

plt.subplot(2, 2, 3)
plt.plot(x, backward)
plt.title('Backward Difference')
plt.xlabel('x')
plt.ylabel("f'(x)")

plt.subplot(2, 2, 4)
plt.plot(x, df_dx)
plt.title('Central Difference')
plt.xlabel('x')
plt.ylabel("f'(x)")

plt.tight_layout()
plt.show()
```
