```python
import numpy as np
import matplotlib.pyplot as plt


def f(x, y):
    return -2 * y

# Define parameters
h_values = np.arange(0.01, 1.01, 0.05)
x_start, x_end = 0, 2

first_x = None
first_y = None


for h in h_values:
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    # Initial condition
    y_values[0] = 1  # y(0) = 1

    # Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])

    
    if first_x is None:
        first_x = x_values.copy()
        first_y = y_values.copy()

   
    first_y_interp = np.interp(x_values, first_x, first_y)
    error = np.abs(y_values - first_y_interp)
    total_error = np.sum(error)

    print("h =", round(h, 2))
    print("Error at each point =", error)
    print("Total accumulated error =", total_error)
    print("                                 ")

    # Plot each step-size run
    plt.plot(x_values, y_values)

plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -2y using Euler's Method")
plt.legend()
plt.show()
```
