```python
import numpy as np
import matplotlib.pyplot as plt
import time

def naive_matrix_multiplication(A, B):
    n = len(A)
    C = [[0] * n for _ in range(n)]
    
    for i in range(n):
        for j in range(n):
            for k in range(n):
                C[i][j] += A[i][k] * B[k][j]
    
    return C

sizes = []
n = 3
for i in range(8):
    sizes.append(n)
    n = n * 2

naive_times = []
numpy_times = []
matrix_numbers = np.arange(1, 9)

for n in sizes:
    A = np.random.rand(n, n)
    B = np.random.rand(n, n)
    start = time.time()
    C_naive = naive_matrix_multiplication(A.tolist(), B.tolist())
    end = time.time()
    naive_times.append(end - start)
    start = time.time()
    C_numpy = np.dot(A, B)
    end = time.time()
    numpy_times.append(end - start)

print("Matrix sizes:", sizes)
print("Naive runtimes:", naive_times)
print("NumPy runtimes:", numpy_times)

plt.figure(figsize=(8,5))
plt.plot(matrix_numbers, naive_times, marker='o', label="Naive")
plt.plot(matrix_numbers, numpy_times, marker='o', label="NumPy")
plt.xlabel("Matrix Number")
plt.ylabel("Runtime in seconds")
plt.title("Naive vs NumPy Matrix Multiplication Runtime")
plt.legend()
plt.grid()
plt.show()
```
