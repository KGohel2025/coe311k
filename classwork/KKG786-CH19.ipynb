```python
import numpy as np
import matplotlib.pyplot as plt


def euler_exact(x):
    return 3 - 0.998*np.exp(-1000*x) - 2.002*np.exp(-x)


def euler_appx(step):
    
    h = step  
    x_start, x_end = 0, 30
    num_steps = int((x_end - x_start) / h)
    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)
    
    
    y_values[0] = 0  
    
    def f(x, y):
        return -1000*y + 3000 - 2000*np.exp(-x)

    
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])
    
    plt.subplot(1, 2, 1)
    plt.plot(x_values, y_values, label=f"Euler's step={step:.2f}")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Euler's Forward")
    
    plt.subplot(1, 2, 2)
    plt.plot(x_values, euler_exact(x_values), 'r', label=f"analytical")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Analytical Solution")
    plt.legend()
    plt.show()
    
    

for i in [.1, 0.01, .001]:
    euler_appx(i)
    

```
