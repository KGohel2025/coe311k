```python
import numpy as np
import matplotlib.pyplot as plt

# Define parameters
h_vals = [0.1, 0.5, 1.0, 1.5]
x_start, x_end = 0, 10

# Function representing dy/dx = -2y
def f(x, y):
    return -2 * y

def exact(x):
    return np.exp(-2*x)

def locale(x, h):
    y_n = exact(x)
    y_step = y_n + h * f(x, y_n)
    y_true = exact(x + h)
    return abs(y_true - y_step)

def globale(y, x):
    return abs(exact(x) - y)

https://jupyter.tacc.cloud/user/kgohel/notebooks/work/ls6/coe311k/classwork/KKG786-CH16.ipynb#for h in h_vals:

    num_steps = int((x_end - x_start) / h)

    x_values = np.linspace(x_start, x_end, num_steps + 1)
    y_values = np.zeros(num_steps + 1)

    y_values[0] = 1  # y(0) = 1

    loc_err = np.zeros(num_steps + 1)
    glob_err = np.zeros(num_steps + 1)

# Implement Euler's method
    for i in range(num_steps):
        y_values[i + 1] = y_values[i] + h * f(x_values[i], y_values[i])

        loc_err[i + 1] = locale(x_values[i], h)
        glob_err[i + 1] = globale(y_values[i + 1], x_values[i + 1])

# Plot the results
    plt.subplot(1,2,1)
    plt.plot(x_values, loc_err, label=f"loc h={h}")
    plt.plot(x_values, glob_err, '--', label=f"glob h={h}")
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title("Approximation of dy/dx = -2y using Euler's Method")
    plt.legend()

    plt.subplot(1,2,2)
    plt.plot(x_values, y_values, label=f"h={h}")

x_exact = np.linspace(x_start, x_end, 400)
plt.subplot(1,2,2)
plt.plot(x_exact, exact(x_exact), 'k', label="exact")

plt.xlabel('x')
plt.ylabel('y')
plt.title("Approximation of dy/dx = -2y using Euler's Method")
plt.legend()

plt.show()
```
