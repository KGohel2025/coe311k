```python
import numpy as np
def DiagnolyDomimant(A):
    row, col = np.shape(A)
    for i in range(row):
        sum = 0
        for j in range(col):
            if(i!=j): 
                sum += np.abs(A[i, j])
            if(sum> np.abs(A[i,i])):
                return False
            
    return True 




A = np.array([[0.582745, 0.48 , 0.10 , 0.  , 0.  ],
              [0.48 , 1.044129, 0.46 , 0.10 , 0.  ],
              [0.10 , 0.46 , 1.10431 , 0.44 , 0.10 ],
              [0.  , 0.10 , 0.44 , 0.963889, 0.42 ],
              [0.  , 0.  , 0.10 , 0.42 , 0.522565]], dtype=float)

print(DiagnolyDomimant(A))


A = np.array([[1, 2, 3, 0, 0],
              [2, 1, 2, 3, 0],
              [3, 2, 1, 2, 3],
              [0, 3, 2, 1, 2],
              [0, 0, 3, 2, 1]])

print(DiagnolyDomimant(A))

```
