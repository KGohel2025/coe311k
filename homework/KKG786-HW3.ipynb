```python
import numpy as np
import matplotlib.pyplot as plt


# calculate taylor series
def taylor(x, term):
    s = 0.0
    for i in range(term):
        s = s+ (((-1)**i) * (x**(2*i + 1)) / np.math.factorial(2*i + 1))
    return s



#arrange x values
x = np.arange(0, 10.0 + 0.1, 0.1)

# list of terms
terms_list = [1, 2, 3, 4]

plt.figure()


for n in terms_list:

    approx_values = []      #stores Taylor approximations
    error_values = []       #stores absolute errors

    # go through each x value
    for i in x:

        # calculate Taylor 
        approx = taylor(i, n)

        # exact sin
        exact = np.sin(i)

        # calculating absolute error
        error = abs(approx - exact)

        # storing values
        approx_values.append(approx)
        error_values.append(error)

    # plot 
    plt.plot(x, error_values, label=n)

# labels
plt.xlabel("x")
plt.ylabel("Absolute Error")
plt.title("Taylor Series Truncation Error for sin(x)")
plt.legend()

# show graph
plt.show()

'''
With more terms the error decreases.  
It does not uniformly decrease. 
It's more accurate with smaller x values rather than larger x values. 


Used AI for creating the calculation of taylor series
why you needed to use an AI : orignally wasn't planning on using AI,
but as I was creating a function for the taylor series I realized there
was probably a function in python that could greatly simplify the code. 
the AI prompt you used: are there any libraries that have functions relating to computing a factorial or Taylor series?
which AI you used: chatgpt

'''


```
