```python
import numpy as np
import math as math
import matplotlib.pyplot as plt
import random as random

#function to approximate pi
def approximatePi(n):
    counter = 0
    for i in range(n):
        #gets your random points
        x = random.uniform(-1,1) 
        y = random.uniform(-1,1) 
        if (x * x + y * y < 1):
               counter = counter + 1
            
    return 4*counter/n

#uses a forloop to plot in steps
for i in range(10, 1000, 20): 
    x= approximatePi(i)
    y = abs(x - math.pi)
    plt.plot(i, y, 'b.')
#plot and labels
plt.xlabel("Run number")
plt.ylabel("Error")
plt.show()
```
