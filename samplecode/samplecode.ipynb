```python
import random 
```

```python
def approximatePi(numsamples):

    numcount = 0
    for i in range (0, numsamples):
        x = random.random()
        y = random.random()
        if (x**2 +y**2 < 1):
            numcount= numcount +1
    return(4*numcount/numsamples)
    
```

```python
print(approximatePi(1000))
```
